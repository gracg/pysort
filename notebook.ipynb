# PySort

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('sorting.csv')
df.head()
```

### Time complexity

```python
strategies = df['name'].unique()
input_sizes = df['size'].unique()

_,ax = plt.subplots()
plt.style.use('fivethirtyeight')
ax.set_xlabel("size")
ax.set_ylabel("total operations")

for strategy_name in strategies:
    strategy = df[df.name == strategy_name]
    ax.plot(input_sizes,strategy['total'],label=strategy_name)

_ = ax.legend()
```

## By operation Type

```python
for name in strategies:
    _,ax = plt.subplots()
    ax.set_xlabel("size")
    ax.set_ylabel("operations")
    
    strategy = df[df.name == name]
    
    ax.plot(input_sizes,strategy['reads'],label='reads')
    ax.plot(input_sizes,strategy['writes'],label='writes')
    ax.plot(input_sizes,strategy['comparisons'],label='comparisons')
    ax.set_title(name)


    _ = ax.legend()
```
